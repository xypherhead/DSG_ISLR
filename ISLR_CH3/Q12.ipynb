```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import math
import random as rn

import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.graphics.regressionplots import *
from sklearn import datasets, linear_model
from patsy import dmatrices
```

The coefficients for both models will same if when plotted, the relationship between x and y is symmetric about y = x.

#### Models which have same coefficient when flipped

```python
rn.seed(1)
x = np.random.normal(0,1,100)
y = -x
lm = sm.OLS(y,x)
res = lm.fit()
print(res.summary())
```

```python
lm2 = sm.OLS(x,y)
res2 = lm2.fit()
print(res2.summary())
```

```python
plt.scatter(x,y)
```

#### Models which don't have same coefficient when flipped

```python
x_diff = np.random.normal(0,1,100)
y_diff = 3*x_diff
lm_diff = sm.OLS(y_diff,x_diff)
res_diff = lm_diff.fit()
print(res_diff.summary())
```

```python
lm_diff2 = sm.OLS(x_diff,y_diff)
res_diff2 = lm_diff2.fit()
print(res_diff2.summary())
```

```python
plt.scatter(y_diff, x_diff)
```
